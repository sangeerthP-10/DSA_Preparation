#Find the second largest element and the second smallest element

```python
def second_largest(arr):
  largest,second_largest=float('-inf'),float('-inf')
  for num in arr:
    if num>largest:
      second_largest=largest
      largest=num
    elif num>second_largest and num<largest:
      second_largest=num
  return second_largest
arr=[1,2,4,3,7,3,8,9]
print(second_largest(arr))
```

```python

def second_smallest(arr):
  smallest,second_smallest=float('inf'),float('inf')
  for num in arr:
    if num<smallest:
      second_smallest=smallest
      smallest=num

    elif num<second_smallest and num>smallest:
      second_smallest=num

  return second_smallest
arr=[2,3,4,1,6,9]
second_smallest_number=second_smallest(arr)
print(f"Second Smallest number is {second_smallest_number}")
```

#Inserting elements in array

```python
def insertBegin(value,arr,N):
  nums=[0]*(N+1)
  nums[0]=value
  for i in range(N):
    nums[i+1]=arr[i]
  print(nums)
def insertEnd(value,arr,N):
  nums=[0]*(N+1)
  nums[N]=value
  for i in range(N):
    nums[i]=arr[i]
  print(nums)



def insertMiddle(value,arr,N,location):
  nums=[0]*(N+1)
  for i in range(location):
    nums[i]=arr[i]

  for i in range(location,N):
    nums[i+1]=arr[i]

  nums[location]=value
  print(nums)


arr=[1,2,3,4]
insertBegin(0,arr,len(arr))
insertEnd(5,arr,len(arr))
insertMiddle(3,arr,len(arr),3)


```

#*****find the maximun product subarray in an array

```python
def max_product_subarray(nums):

    res = max_prod = min_prod = nums[0]

    for i in range(1, len(nums)):
        curr = nums[i]

        # If the number is negative, swapping max and min
        # accounts for the sign flip when multiplied
        if curr < 0:
            max_prod, min_prod = min_prod, max_prod

        # Update current max and min product
        max_prod = max(curr, max_prod * curr)
        min_prod = min(curr, min_prod * curr)

        # Keep track of the overall maximum found so far
        res = max(res, max_prod)

    return res

# Example Usage:
nums = [2, 3, -2, 4]
print(f"Maximum Product: {max_product_subarray(nums)}") # Output: 6
```

```python
arr = [1, 2, 3, -3, -4,]
res = max_prod = min_prod = arr[0]

for i in range(1, len(arr)):
    x = arr[i]
    if x < 0:
        max_prod, min_prod = min_prod, max_prod

    max_prod = max(x, max_prod * x)
    min_prod = min(x, min_prod * x)
    res = max(res, max_prod)

print(res)
```

#Find all symmetric pairs in the array of pairs

#*******sort elemnts of an array by frequency


```python
arr=[1,2,1,3,1,4,5,6,5]
sorted_arr=dict()
for num in arr:
  if num in sorted_arr:
    sorted_arr[num]+=1
  else:
    sorted_arr[num]=1
print(sorted_arr)


result=sorted(arr,key=lambda x:(-sorted_arr[x],x))
print(result)
```
