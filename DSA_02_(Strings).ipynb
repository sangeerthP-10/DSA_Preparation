#Check the string is palindrome

```python
 str="ABCCBA"
if str==str[::-1]:
  print("Palindrome")
else:
  print("Not palindrome")
```

```python
def _ispalindrome(s):
  start,end=0,len(s)-1
  while start<=end:
    if s[start]!=s[end]:
      return "Not Palindrome"
    else:
      return "Its a palindrome"

str=input()
print(_ispalindrome(str))
```

#Count the number of vowels,consonants and spaces

```python
def count_character(str):
  vowels=consonants=spaces=0
  for char in str:
    if char.lower() in 'aeiou':
      vowels+=1
    elif char.isalpha():
      consonants+=1
    elif char.isspace():
      spaces+=1

  return vowels,consonants,spaces


str=input()
vowels,consonants,spaces=count_character(str)
print("Vowels: ",vowels)
print("consonants: ",consonants)
print("Spaces :",spaces)
```

#Remove all vowels from the string

```python
def remove_vowels(str):
  result=""
  for char in str:
    if char.lower() not in 'aeiou':
      result+=char
  return result

str=input()
print("New string: ",remove_vowels(str))
```

```python
def remove_vowelss(str):
  vowels="aeiou"
  return "".join(char for char in str if char.lower() not in vowels)

str=input()

print("New string: ",remove_vowels(str))
```

#Remove spaces from the string

```python
def remove_spaces(str):

  return "".join(char for char in str if char.isspace()==False)

str=input()

print("New string: ",remove_spaces(str))
```

```python
def remove_spaces(str):

  return str.replace(" ","")

str=input()

print("New string: ",remove_spaces(str))
```

#Remove  characters from a string except alphabets

```python
def remove_chars(str):
  return "".join(char for char in str if char.isalpha()==True)

str=input()
print(remove_chars(str))

```

#Capitalize first and last character of each word of a string

```python
str=input()
result=""
for i in range(len(str)):
  space_before= (i==0 or (i>0 and str[i-1]==" "))
  space_after= (i==(len(str)-1) or (i<len(str) and str[i+1]==" "))
  if space_after or space_before:
    result+=str[i].upper()
    continue
  result+=str[i]

result

```

#Calculate the frequency of character in a string

```python
str=input()
freq=dict()
for char in str:
  if char in freq:
    freq[char]+=1
  else:
    freq[char]=1
arr=[]
for key,value in freq.items():
  a=f'{key}{value}'
  arr.append(a)

arr.sort()
for i in arr:
  print(i,end=" ")

```
